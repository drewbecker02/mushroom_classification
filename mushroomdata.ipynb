```python
from ast import fix_missing_locations
"""This program is going to give us a usable mushroom dataset for classification
by adding family and species to the csv files."""

primary_data = 'primary_data.txt'
seconday_data = 'secondary_data_generated.txt'

species_list = []
family_list =[]
matched_species =[[]]
data_list = []
#add to species and family lists
with open(primary_data) as f:
  lines = f.readlines()
  for line in lines:
    line=line.split(';')
    print (line[0])
    species = "'"+line[1]+"'"
    species_list += [species]
    if line[0] not in family_list:
      family = line[0]
      family_list += [family]
      print (family_list.index(family))
      matched_species+=[[]]
    matched_species[family_list.index(family)] += [species]
print (species_list)
print (family_list)
print (matched_species)


with open(seconday_data) as f:
  lines =f.readlines()
  index = 0
  count =0
  num_per_species = 353
  i = 0
  for line in lines:
    line = line.rstrip('\n')
    line=line.split(';')
    line += [species_list[index]]
    for fam in matched_species:
      if line[len(line)-1] in fam:
        line+=[family_list[matched_species.index(fam)]]
    if i>0 and i % 353 == 0:
      index +=1
    i+=1
    datastring = ', '.join(line)
    data_list.append(datastring)
with open('data_families.txt', 'w') as f:
  for line in data_list:
    f.writelines(line)
    f.writelines('\n')

    
    
```
